```python
import requests
import pandas as pd
from IPython.display import display
from google.colab import files

# 讓使用者輸入 API 金鑰
api_key = input('請輸入你的 Google Maps API 金鑰: ')

# 搜尋地點的基本資訊
def search_place(query):
    endpoint = 'https://maps.googleapis.com/maps/api/place/textsearch/json'
    params = {
        'query': query,
        'key': api_key,
        'language': 'zh-TW'  # 設定語言為繁體中文
    }

    response = requests.get(endpoint, params=params)
    results = response.json()['results']

    # 整理資料到 Pandas DataFrame
    stores_info = []
    for result in results:
        name = result['name']
        address = result['formatted_address']
        rating = result.get('rating', '')  # 若無評分則為空字串
        place_id = result['place_id']
        link = f'https://www.google.com/maps/place/?q=place_id:{place_id}'  # 地點連結

        store_info = {
            '名稱': name,
            '地址': address,
            '星數': rating,
            '地點連結': link
        }
        stores_info.append(store_info)

    # 轉換成 DataFrame
    df = pd.DataFrame(stores_info)

    # 顯示 DataFrame
    display(df)

    # 儲存成 CSV 檔案
    df.to_csv('places.csv', index=False, encoding='utf-8-sig')
    print('已將搜尋結果儲存為 places.csv')

    # 下載 CSV 檔案
    files.download('places.csv')

# 讓使用者輸入想搜尋的地點
user_input = input('請輸入你想搜尋的地點或餐廳名稱: ')

# 呼叫搜尋地點函式
search_place(user_input)
```
